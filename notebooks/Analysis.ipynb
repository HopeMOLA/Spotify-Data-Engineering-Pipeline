***Top 10 Artists Analysis***

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../data/processed/clean_spotify_data.csv")

top_artists = df['artist'].value_counts().head(10)

print(top_artists)
```

```python
top_artists.plot(kind='bar')

plt.title("Top 10 Artists in Playlist")
plt.xlabel("Artist")
plt.ylabel("Number of Songs")

plt.show()
```

Popularity Distribution

```python
plt.hist(df['popularity'], bins=10)

plt.title("Popularity Distribution")
plt.xlabel("Popularity Score")
plt.ylabel("Number of Songs")

plt.show()
```

***Average Popularity by Artis***

```python
avg_popularity = df.groupby('artist')['popularity'].mean().sort_values(ascending=False)

print(avg_popularity.head(10))
```

***Most Popular Songs***

```python
top_songs = df.sort_values(by='popularity', ascending=False).head(10)

print(top_songs)
```
